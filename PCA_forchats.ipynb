```python
# Data cleaning for df_chat
# dropping any row with null values
df_chat.dropna(inplace=True)

# type conversion
strColumns = ["account_id", "alliance_id", "raw_message", "filtered_message", "filtered_content", "filter_detected_language"]
intColumns = ["filtered", "is_family_friendly"]
floatColumns = ["risk", "GENERAL_RISK", "BULLYING", "VIOLENCE", "RELATIONSHIP_SEXUAL_CONTENT", "VULGARITY", "DRUGS_ALCOHOL", "IN_APP", "ALARM", "FRAUD", "HATE_SPEECH", "RELIGIOUS", "WEBSITE", "CHILD_GROOMING", "PUBLIC_THREAT", "EXTREMISM", "SUBVERSIVE", "SENTIMENT", "POLITICS"]
timeColumns = ["timestamp", "date"]

for colName in intColumns:
  df_chat[colName] = pd.to_numeric(df_chat[colName], errors='coerce')

for colName in floatColumns:
  df_chat[colName] = pd.to_numeric(df_chat[colName], errors='coerce')

for colName in timeColumns:
  df_chat[colName] = pd.to_datetime(df_chat[colName], errors='coerce')

df_chat.dropna(inplace=True)

df_chat.info()
```

```python
#df only contains columns/features which are numerical 

df=df_chat.select_dtypes(include='float64')
filtered=df_chat.select_dtypes(include='float64')
df.head()
```

```python
# Import PCA Algorithm
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

#scale values to same range 
scaler=StandardScaler()
scaler.fit(df)
df=scaler.transform(df)

# Initialize the algorithm and set the number of PC's
pca = PCA(n_components=3)

# Fit the model to data 
pca.fit(df)
# Get list of PC's
pca.components_
# Transform the model to data 
pca.transform(df)
print(pca.components_.shape)


print(df.shape)
print("var explained by 3 pc = ",sum(pca.explained_variance_ratio_ *100))

import numpy as np
import matplotlib.pyplot as plt

plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of components')
plt.ylabel('Explained variance')
```
